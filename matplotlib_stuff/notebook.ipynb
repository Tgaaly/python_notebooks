```python
import numpy as np
import matplotlib.pyplot as plt


data= {'cyclist': 14,
'golf-cart': 8935,
'large-vehicle': 251,
'motorcyclist': 14,
'pedestrian': 1567,
'pedestrian-with-object': 40,
'unknown/other': 1560,
'vehicle': 35672}

classes = data.keys()
x_pos = np.arange(len(classes))
# counts = [np.log(float(d)) for i, d in enumerate(data.values())]
counts = data.values()

# fig = plt.figure(figsize=(8, 9))
plt.bar(x_pos, counts, align='center', alpha=0.5, color='blue')
plt.xticks(x_pos, classes, rotation='vertical')
plt.ylabel('Number of Objects')
plt.title('Frequencies of Taxonomy')
plt.ylim(0, 6000)

 
plt.show()
```
